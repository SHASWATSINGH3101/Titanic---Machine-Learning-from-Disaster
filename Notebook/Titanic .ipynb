```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
```

```python
def random_sampling_imputation(df, column):
    # Get non-null values from the column
    non_null_values = df[column].dropna().values
    
    # Count null values
    null_count = df[column].isnull().sum()
    
    # Randomly sample from non-null values
    random_sample = np.random.choice(non_null_values, size=null_count, replace=True)
    
    # Create a copy of the original column
    imputed_column = df[column].copy()
    
    # Fill null values with the random sample
    imputed_column.loc[imputed_column.isnull()] = random_sample
    
    return imputed_column
```

```python
X['journey'] = random_sampling_imputation(X, 'Embarked')
```

```python
journey_column = X[['journey']]  # Make sure to use double brackets to create a DataFrame

# Perform one-hot encoding
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_journey = ohe.fit_transform(journey_column)

# Create new column names
new_column_names = [f'Journey_{cat}' for cat in ohe.categories_[0]]

# Create a new DataFrame with the encoded values
encoded_df = pd.DataFrame(encoded_journey, columns=new_column_names, index=X.index)

# Concatenate the new columns to the original DataFrame
X = pd.concat([X, encoded_df], axis=1)
```
